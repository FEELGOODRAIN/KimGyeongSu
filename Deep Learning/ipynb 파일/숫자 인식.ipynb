2. 모델 데이터를 다운받는다.
 - 이미지를 생성한다.

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
from PIL import Image
import sys
image = Image.open('image3.bmp')
model = load_model('MNIST_CNN.hdf5')
def predict(image):
    if image.mode == '1':
        image = image.convert('L')
    x = np.array(image)
    x = x.astype('float32') / 255
    image_array = np.expand_dims(x, axis=-1)
    input_data = np.expand_dims(image_array, axis=0)
    plt.imshow(image_array , cmap = 'gray')
    plt.show()
    print(model.predict(input_data)[0].tolist())
    print('예측한 숫자 : {}'.format(model.predict(input_data)[0].tolist().index(max(model.predict(input_data)[0].tolist()))))
predict(image)
```
