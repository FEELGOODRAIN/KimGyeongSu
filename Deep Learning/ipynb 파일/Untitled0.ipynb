```python
def solution(nums):
    answer = 0
    result = []
    for i in nums:
        for j in nums:
            for k in nums:
                result.append(i+j+k)


    nums = sorted(nums)
    a = list(set(result))

    a = a[:a.index(sum(nums[-3:]))+1]


    a = a[a.index(sum(nums[:3]))-1:]

    for i in a:
        b = 0
        for j in range(1,i+1):
            if i%j==0:
                b +=1
        if b == 2:
            answer+=1

    return answer
```

```python
def solution(arr1, arr2):

    answer = [[0] * len(arr1[0]) for _ in range(len(arr1))]
    for i in range(len(arr1)):
        for j in range(len(arr1[i])):
            answer[i][j] = arr1[i][j] + arr2[i][j]
    return answer
```
