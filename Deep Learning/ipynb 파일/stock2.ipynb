```python
dow = fdr.DataReader('DJI' , startdate , enddate)
nas = fdr.DataReader('IXIC' , startdate , enddate)
nvda = fdr.DataReader('NVDA' , startdate , enddate)
amd = fdr.DataReader('AMD' , startdate , enddate)
intc = fdr.DataReader('INTC' , startdate , enddate)

kospi = fdr.DataReader('KS11' , startdate , enddate)
```

```python
stock['dow'] = dow['Close'].shift(1)
stock['nas'] = nas['Close'].shift(1)
stock['nvda'] = nvda['Close'].shift(1)
stock['amd'] = amd['Close'].shift(1)
stock['intc'] = intc['Close'].shift(1)
stock['kospi'] = kospi['Close']
```

```python
stock.columns
```

```python
stock = stock[['Open', 'High', 'Low', 'dow', 'nas', 'nvda', 'amd',
       'intc', 'kospi' , 'Close']]
```

```python
stock
```

```python
stock.dropna(axis = 0 , inplace = True)
stock
```

```python
from sklearn.preprocessing import StandardScaler , MinMaxScaler

scaler = StandardScaler()

stock.iloc[:,:-1] = scaler.fit_transform(stock.iloc[:,:-1])
```

```python
stock
```

```python
import pandas as pd
import numpy as np
```

```python
window_size = 20
x = []
y = []

for i in range(len(stock) - window_size):
  temp = stock.iloc[i:i+window_size , :-1]
  x.append(temp)
  temp1 = stock.iloc[i+window_size , -1]
  y.append(temp1)


x = np.array(x)
y = np.array(y)


print(x.shape , y.shape)
```

```python
begin_2023 = stock.index.get_loc('2023-01-03')
begin_2023
```

```python
split_point = begin_2023 - window_size
```

```python
x_train = x[:split_point]
y_train = y[:split_point]

x_test = x[split_point:]
y_test = y[split_point:]

print(x_train.shape , y_train.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_val , y_train , y_val = train_test_split(x_train , y_train , test_size = 0.2 , shuffle = True)
```

```python
print(x_train.shape , y_train.shape , '\n' , x_val.shape , y_val.shape , '\n' , x_test.shape , y_test.shape)
```

```python
from tensorflow.keras.models import Sequential , load_model
from tensorflow.keras.layers import LSTM , Dense , Dropout , Flatten
from tensorflow.keras.callbacks import ModelCheckpoint , EarlyStopping
```

```python
mc = ModelCheckpoint('stock1.h5' , monitor = 'val_mae' , save_best_only = True , verbose = 1)
ea = EarlyStopping(monitor = 'val_loss' , mode = 'min' , verbose = 1 , patience = 20)
```

```python
model = Sequential()
model.add(LSTM(32 , activation = 'tanh' , input_shape = (x_train.shape[1] , x_train.shape[2])))
model.add(Dropout(0.3))
model.add(Dense(64 , activation = 'relu'))
model.add(Dropout(0.3))
model.add(Dense(1))

model.compile(loss = 'mae' , optimizer = 'adam' , metrics = 'mae')
```

```python
history = model.fit(x_train , y_train , epochs = 2000 , callbacks = [mc , ea] , validation_data = (x_val , y_val))
```

```python
score = model.evaluate(x_test , y_test , verbose = 0)
print('test loss : ' , score[0])
print(score[1])
```

```python
import matplotlib.pyplot as plt
y_pred = model.predict(x_test)
days_2023 = stock.index[stock.index >= '2023-01-02']

plt.figure(figsize = (12,8))
plt.title('2023')
plt.plot(days_2023 , y_test , 'orange' , label = 'truth')
plt.plot(days_2023 , y_pred , 'skyblue' , label = 'prediction')
plt.legend()
plt.show()
```

```python
def execute(company):
     # Press Ctrl+F8 to toggle the breakpoint.
    stock = fdr.DataReader(company , '2020-01-01' , ' 2023-11-28')
    stock['Close2'] = stock['Close']
    stock.drop(['Close','Change'] , axis = 1 , inplace = True)
    scaler = StandardScaler()
    stock.iloc[:,:-1] = scaler.fit_transform(stock.iloc[:,:-1])
    window_size = 20
    x = []
    y = []

    for i in range(len(stock) - window_size):
        temp = stock.iloc[i:i + window_size, :-1]
        x.append(temp)
        temp1 = stock.iloc[i + window_size, -1]
        y.append(temp1)

    x = np.array(x)
    y = np.array(y)

    begin_2023 = stock.index.get_loc('2023-01-02')
    split_point = begin_2023 - window_size
    x_train = x[:split_point]
    y_train = y[:split_point]

    x_test = x[split_point:]
    y_test = y[split_point:]
    x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, shuffle=True)

    mc = ModelCheckpoint('stock1.h5', monitor='val_mae', save_best_only=True, verbose=1)
    ea = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=20)
    model = Sequential()
    model.add(LSTM(32, activation='tanh', input_shape=(x_train.shape[1], x_train.shape[2])))
    model.add(Dropout(0.3))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.3))
    model.add(Dense(1))

    model.compile(loss='mae', optimizer='adam', metrics='mae')
    history = model.fit(x_train, y_train, epochs=2000, callbacks=[mc, ea], validation_data=(x_val, y_val))
    score = model.evaluate(x_test, y_test, verbose=0)
    print('test loss : ', score[0])
    print(score[1])
```
